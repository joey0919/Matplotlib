```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4 ,8]
plt.plot(x, y)
plt.title('꺽은선 그래프', fontdict={'family': 'HYGungSo-Bold', 'size':20}) # 개별 폰트 설정
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red')
plt.ylabel('Y축', color='#00aabb')
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red', loc='right') # left, center, right
plt.ylabel('Y축', color='#00aabb', loc='top') # top, center, bottom
plt.show()
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
```
